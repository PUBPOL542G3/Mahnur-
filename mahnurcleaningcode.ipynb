In this jupyter notebook I will use the python code to upload and clean an .xlsx file from github.

This code imports and reads the pandas library so that I can upload and manipualte files throughout this code. 

```python
 import pandas as pd
```

Here, I am importing,reading and dispalying the raw CO2 emissions data from my Githib repository "PUBPOL542G3/Mahnur". I display the data in order to double check that all the data was imported correctly.

```python
Mahnur_file_location = 'https://github.com/PUBPOL542G3/Mahnur-/raw/main/Mahnur%20Raw.xlsx'
DF_C02 = pd.read_excel(Mahnur_file_location)
display(DF_C02)
```

```python
DF_C02.columns
```

After importing the data, the following code narrows down the data that I imported to only show me what I am looking for. In this case, I refined the full data set to only display the country and the corresponding CO2 emissions from the year 2015. I display the data frame after these changes to make sure everything worked properly.

```python
refined_columns = ['country', 2015]
```

```python
DF_C02 = DF_C02 [refined_columns]
```

```python
DF_C02
```

Next, I renamed the column previously labeled as 2015 to "CO2". This is so that when I merge my data with my groupmembers 2015 data, we will know what the data represents. I display the renamed column to make certain that everything worked properly.

```python
DF_C02.rename (columns={ 2015 :'CO2'}, inplace=True)
```

```python
DF_C02
```

After renaming the column, I needed to rename a few of the countries in order to make sure that they matched the naming conventions of my groupmates. In this code, I identified which existing country names I wanted to change and then presented the new country name that I wanted to replace the old name with.

```python
DF_C02_revised = DF_C02.replace ({'Congo, Dem. Rep.' : 'Democratic Republic of the Congo','Congo, Rep.':'Republic of the Congo',
'Kyrgyz Republic':'Kyrgyzstan','Slovak Republic':'Slovakia'})
```

```python
DF_C02_revised [DF_C02_revised['country']== 'Democratic Republic of the Congo']
```

I am downloading the "DF_C02_revised" data frame onto my desktop as an excel file named "mahnurcleanedfile". I will then take this excel document and upload it to the team's Github repository for further use.
